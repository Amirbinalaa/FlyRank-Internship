```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(f"🎉 Successfully loaded! Data shape: {df.shape}")
```

```python
filtered_df = df[(df["impressions_90d"] > 0) & (df["content_age_days"] >= 90)].drop_duplicates(subset=["content_id"])

#Calculating the 3 numbers that support my lane
total_pages = len(filtered_df)
declining_pages = filtered_df[filtered_df["trend_direction"] == "down"]
pct_declining = (len(declining_pages) / total_pages) * 100

stale_pages = filtered_df[filtered_df["content_age_days"] >= 180]
pct_stale = (len(stale_pages) / total_pages) * 100

print(f"Total Eligible Pages in Starter Set: {total_pages}")
print(f"Pages with Declining Trend: {len(declining_pages)} ({pct_declining:.1f}%)")
print(f"Stale Pages (Age >= 180 days): {len(stale_pages)} ({pct_stale:.1f}%)")
```
